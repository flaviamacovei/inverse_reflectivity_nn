```python
import yaml

materials_file = "../../material_embedding/materials.yaml"
with open(materials_file) as f:
    contents = yaml.safe_load(f)
```

```python
import os
_, _, material_filenames = next(os.walk('.'))
```

```python
for material in material_filenames:
    if not material.endswith('.csv'):
        continue
    title = material.split('.')[0]
    print(f"M_{title}")
```

```python
with open(materials_file, 'w') as f:
    yaml.dump(contents, f)
```
